# Data Understanding

## Sumber Data

Dataset diambil dari link dibawh ini:

https://huggingface.co/datasets/farhamu/tokopedia-product-reviews-2019

Dataset tokopedia-product0reviews-2019 merupakan kumpulan data review pelanggan dari produk pada masing-masing kategori. Kumpulan data ini berisi 40.607 ulasan produk dari Tokopedia, salah satu platform e-commerce terbesar di Indonesia, yang dikumpulkan pada tahun 2019. Kumpulan data ini memberikan wawasan berharga tentang sentimen pelanggan dan perilaku belanja di pasar e-commerce Indonesia.

## Pengambilan data

Mengambil data pada site Hugging Face, telah disediakan berbagai cara unutuk mengakses dataset di antaranya melalui library python seperti Dataset, Pandas, Croiisant, dan Pollars. Berikut cara mengambil data dan menampilkan data yang diambil dari sites Hugging Face:

```python
from datasets import load_dataset
import pandas as pd

# Ambil dataset dari Hugging Face
dataset = load_dataset("farhamu/tokopedia-product-reviews-2019")

df = dataset["train"].to_pandas()

print(display(df.head()))
```

### Fitur dalam Dataset

```python
df.columns
```

Berdasarkan data di atas fitur yang terdapat pada dataset tersebut yakni:

1.   text – Ulasan pelanggan dalam bentuk teks.
2.   rating – Penilaian numerik pelanggan (biasanya 1–5).
3.   category – Kategori produk yang diulas (misal: Elektronik, Fashion).
4.   product_name – Nama produk yang diulas.
5.   product_id – ID unik produk di Tokopedia.
6.   sold – Jumlah produk yang telah terjual.
7.   shop_id – ID unik toko penjual produk.
8.   product_url – URL halaman produk di Tokopedia.
