# Bussiness Understanding Tokepedia

Tokopedia merupakan salah satu perusahaan e-commerce terbesar di Indonesia yang berdiri pada tahun 2009. Perusahaan ini didirikan oleh William Tanuwijaya dan Leontinus Alpha Edison dengan visi untuk “mendemokratisasi perdagangan melalui teknologi” dan memberdayakan pelaku usaha mikro, kecil, dan menengah (UMKM) agar dapat menjangkau pasar secara lebih luas melalui kanal digital.

## Latar Belakang

Perusahan Tokopedia menyediakan berbagai produk dengan beberapa kategori di antaranya fashion, elektronik, kebutuhan rumah tangga, layanan digital, dll. Dengan adanya traksaksi setiap harinya, data ulasan pelanggan digunakan untuk mengevaluasi kualitas pelayanan, produk, serta pengalaman kustomer dalam menggunakan Tokopedia.

Dataset yang digunakan dalam proyek ini berisi lebih dari 40.000 record data ulasan dari pelanggan. Data ini dapat dimanfaatkan sebagai bahan analisis tingkat kepuasan dari pelanggan, indentifikasi pola ulasan, serta memberikan rekomendasi untuk meningkatkan kualitas dan penjualan.

## Tujuan Bisnis

- **Mengukur tingkat kepuasan pelanggan melalui ulasan atau rating**:
Dengan menggunakan fitur ulasan atau rating dari dataset kita dapat mengetahui tingkat kepuasan dari pelanggan yang dapat dijadikan sebagai acuan dalam meningkatkan kualitas pelayanan.
- **Menganalisis hubungan antara setiap fitur untuk memahami perilaku pembelian**: Dengan melakukan analisa terhadap setiap fitur pada perilaku pembelian dapat mengetahui fitur yang dapat mempengaruhi pembeli dalam mengambil keputusan dalam penjualan produk.

## Stakeholder


- **Manajemen Tokopedia**: Manajemen dapat mengevaluasi peforma dari layanan dan produk sebagai acuan untuk pengambilan keputusan dalam keberlangsungan perusahaan.
- **Tim Pemasaran**: Untuk menentukan strategi promosi yang tepat pada produk dengan rating terendah.
- **Seller**: Dengan adanya Bussiness Understanding Seller dapat meningktakan kualitas dari produk berdasarkan rating maupun ulasan.


## Nilai Bisnis

- Meningkatkan kepuasan pelanggan melalui perbaikan kualitas produk dan layanan berdasarkan data ulasan ataupun rating.

- Mengoptimalkan strategi promosi dengan mengidentifikasi produk yang berpotensi untuk ditingkatkan performanya.
